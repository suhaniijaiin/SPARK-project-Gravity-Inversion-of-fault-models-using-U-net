```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .dat file (assuming it is just a list of numbers in text format)
data = np.loadtxt('dataa/tra&val/data1.dat')  # shape should be (1024,)


```

```python
# Reshape to 32x32
density_map = data.reshape((32, 32))

# Plot with color map
plt.imshow(density_map, cmap='viridis')  # You can change cmap as needed
plt.colorbar(label='Density')
plt.title('gravity data(32x32)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

# Load the .mat file
mat = scipy.io.loadmat('200model/tra&val/model67.mat')

# Extract and reshape d

d = mat['d'].reshape((32, 32))  # 2D density map

# Extract and reshape m
m = mat['m'].reshape((16, 32, 32))  # 3D model volume
```

```python
plt.imshow(d, cmap='viridis')
plt.colorbar(label='d value')
plt.title(' gravity data (32x32)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
m = mat['m'].reshape((16, 32, 32))  # reshape m from (16384, 1) to (16, 32, 32)

# Create binary mask where m == 1000
mask = m == 1000
mask0= m==0

# Create a voxel plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.voxels(mask, facecolors='red', edgecolor='k')
ax.voxels(mask0, facecolors='yellow', edgecolor='k')
ax.view_init(elev=30, azim=-90)


ax.set_title('3D Fault Structure (m)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.tight_layout()
plt.show()
```
